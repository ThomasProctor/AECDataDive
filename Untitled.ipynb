```python
from __future__ import division
import pandas as pd
```

```python
with open('datapath.txt', 'r') as f:
    datapath = f.readlines()[0].strip()
```

```python
cyf_indiv = pd.read_csv(datapath
                        + 'cyf-individual-episodes.csv')
```

```python
cyf_indiv.set_index('client_id', inplace=True)
```

```python
%ls $datapath
```

```python
race_dict = pd.read_csv(datapath
                        + 'RaceReCategorizations_BehavioralDataset.csv')
```

```python
cyf_indiv
```

```python
race_dict.set_index('OriginalRace', inplace=True)
```

```python
race_dict = race_dict['SimplifiedRace']
```

```python
race_dict.unique()
```

```python
new_race = ['Mixed Race',
 'Mixed Race',
 'Asian or Pacific Islander',
 'Asian or Pacific Islander',
 'Asian or Pacific Islander',
 'Asian or Pacific Islander',
 'Mixed Race',
 'Asian or Pacific Islander',
 'Mixed Race',
 'Mixed Race',
 'Mixed Race',
 'Mixed Race',
 'Asian or Pacific Islander',
 'Mixed Race',
'Mixed Race',
 'Mixed Race',
 'Mixed Race',
 'Mixed Race']
```

```python
len(new_race)
```

```python
uncoded = cyf_indiv[~cyf_indiv['RACE'].isin(race_dict.index.tolist())]['RACE'].unique().tolist()
```

```python
race_dict = race_dict.append(pd.Series({i[0]:i[1] for i in zip(uncoded, new_race)}))
```

```python
cyf_indiv.shape
```

```python
cyf_indiv['SIMPLIFIED_RACE'] = race_dict[cyf_indiv['RACE'].fillna('NA')].values
```

```python
cyf_indiv
```

```python
cyf_indiv.to_csv(datapath + 'cyf_individuals_simplified_race.csv')
```

```python
all_mci = pd.read_csv('MCI_IDS_in_all_systems.csv', header=None)[0].tolist()
```

```python
all_systems = cyf_indiv.loc[all_mci]
```

```python
all_systems['SIMPLIFIED_RACE'].value_counts() / all_systems.shape[0]
```

```python
all_systems['BRTH_DT'] = pd.to_datetime(all_systems['BRTH_DT'])
```

```python
early_birth_date = all_systems['BRTH_DT'].min()
```

```python
dt_birth_dates = pd.to_datetime(cyf_indiv['BRTH_DT'], errors='coerce')
young_cyf = cyf_indiv[dt_birth_dates > early_birth_date]
```

```python
young_age = (pd.datetime.today() - pd.to_datetime(young_cyf['BRTH_DT'])).dt.days / 365
```

```python
all_age = (pd.datetime.today() - all_systems['BRTH_DT']).dt.days / 365
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist([young_age, all_age], alpha=0.4, normed=True, label=['CYF Only', 'All Services'])
plt.xlabel('age')
plt.ylabel('frequency (ratio of users)')
plt.legend(loc='upper right')
plt.title('Age breakdown of users by service usage')
plt.savefig('Age breakdown of users by service usage.png')
plt.show()
```
